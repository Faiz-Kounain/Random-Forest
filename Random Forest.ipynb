```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Fraud_check.csv')
df.head()
```

```python
#renaming the columns
df.rename(columns={'Marital.Status':'Marital_Status','Taxable.Income':'Taxable_Income',
                      'City.Population':'City_Population','Work.Experience':'Work_Experience'},inplace=True)
df.head()
```

```python
#converting independent categorical variable  to numerial columns
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
a=pd.Series(df['Taxable_Income'])
```

```python
# If income is less than 30000 label is good else risky 
lst=[]
for i in a:
    x='Good' if i<=30000 else 'Risky'
    lst.append(x)
print(lst)
```

```python
a=pd.DataFrame(lst)
```

```python
df['Tax_Inc']=a
df.head()
```

```python
df.columns
```

```python
# encoding target variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Tax_Inc']=le.fit_transform(df['Tax_Inc'])
df.head()
```

```python
#Normalizing
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
df=mm.fit_transform(df)
```

```python
df
```

```python
# again converting to dataframe
df=pd.DataFrame(df,columns=['Taxable_Income', 'City_Population', 'Work_Experience', 'Undergrad_YES',
       'Marital_Status_Married', 'Marital_Status_Single', 'Urban_YES',
       'Tax_Inc'])
```

```python
df.head()
```

```python
df=df.drop(columns=['Taxable_Income'],axis=1)
df.head()
```

```python
#X&y
X=df.drop(columns=['Tax_Inc'],axis=1)
y=df['Tax_Inc']
```

```python
# Splitting data into training and testing data set
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=40)
```

```python
#import classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
help(RandomForestClassifier)
```

```python
model=RandomForestClassifier(n_estimators=100,max_features=3,random_state=100)
```

```python
#fit the model
model.fit(X_train,y_train)
```

```python
ypred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ypred,y_test)
```

```python
#train Accuracy
model.score(X_train,y_train)  # by default parameter 100% accuracy as max_depth=None
```

```python
#test Accuracy
model.score(X_test,y_test)
```

```python
# important features
model.feature_importances_
```

```python
# grid Search cv
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[30,60,100,150,250,500],'max_features':[2,3,4]}
grid=GridSearchCV(estimator=RandomForestClassifier(),param_grid=param_grid,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```
